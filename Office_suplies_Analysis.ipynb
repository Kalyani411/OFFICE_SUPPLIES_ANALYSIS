<div style="text-align: left; background-color:black; padding: 5px; color: white; font-size: 18px; margin-top: 10px;">
    
What are the total sales for each product category?
</div>

 



```python
total_sales_by_category = df.groupby('Item')['Total Sales'].sum().reset_index()
total_sales_by_category
```

<div style="text-align: left; background-color:black; padding: 5px; color: white; font-size: 18px; margin-top: 10px;">
    
Which product category has the highest sales?
</div>

 




```python
highest_sales_category = total_sales_by_category.loc[total_sales_by_category['Total Sales'].idxmax()]
print(highest_sales_category)
```

<div style="text-align: left; background-color:black; padding: 5px; color: white; font-size: 18px; margin-top: 10px;">
    
Identify the top 10 best-selling products?
</div>

 




```python
top_10_best_selling_products = df.sort_values(by='Total Sales', ascending=False).head(10)
top_10_best_selling_products
```

<div style="text-align: left; background-color:black; padding: 5px; color: white; font-size: 18px; margin-top: 10px;">
    
VISUALIZATION
</div>

 




```python
plt.figure(figsize=(12, 8))
plt.barh(top_10_best_selling_products['Rep'], top_10_best_selling_products['Total Sales'], color='skyblue')
plt.xlabel('Total Sales')
plt.ylabel('Customer')
plt.title('Top 10 Customers by Sales')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest sales on top
plt.show()
```

<div style="text-align: left; background-color:black; padding: 5px; color: white; font-size: 18px; margin-top: 10px;">
    
Analyze customer purchase frequency.
</div>

 




```python
purchase_frequency = df.groupby('Rep').size().reset_index(name='Purchase Frequency')
purchase_frequency
```

<div style="text-align: left; background-color:black; padding: 5px; color: white; font-size: 18px; margin-top: 10px;">
    
VISUALIZATION
</div>

 




```python

plt.figure(figsize=(12, 8))
bars = plt.barh(purchase_frequency['Rep'], purchase_frequency['Purchase Frequency'], color='skyblue')
plt.xlabel('Purchase Frequency')
plt.ylabel('Customer')
plt.title('Customer Purchase Frequency')

# Add the values on top of the bars
for bar in bars:
    plt.text(
        bar.get_width(), 
        bar.get_y() + bar.get_height() / 2,
        str(int(bar.get_width())),
        va='center',
        ha='left',
        color='black'
    )

plt.show()
```

<div style="text-align: left; background-color:green; padding: 5px; color: white; font-size: 18px; margin-top: 10px;">
    
Extract the year and month from the Orderdate column
</div>

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['YearMonth'] = df['OrderDate'].dt.to_period('M')
df
```

```python
last_year = df['OrderDate'].max() - pd.DateOffset(years=1)
df_last_year = df[df['OrderDate'] >= last_year]
df_last_year
```

<div style="text-align: left; background-color:black; padding: 5px; color: white; font-size: 18px; margin-top: 10px;">
    
What are the monthly sales trends over the past year?
</div>

 




```python
# Filter data for the past year
last_year = df['OrderDate'].max() - pd.DateOffset(years=1)
df_last_year = df[df['OrderDate'] >= last_year]

monthly_sales = df_last_year.groupby('YearMonth')['Total Sales'].sum().reset_index()

# Convert "YearMonth" back to datetime for plotting
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].dt.to_timestamp()

# Plot the monthly sales trends
plt.figure(figsize=(12, 8))
plt.plot(monthly_sales['YearMonth'], monthly_sales['Total Sales'], marker='o', linestyle='-')
# Annotate the values on the plot
for i in range(len(monthly_sales)):
    plt.text(
        monthly_sales['YearMonth'][i], 
        monthly_sales['Total Sales'][i], 
        str(round(monthly_sales['Total Sales'][i], 2)),
        ha='right', 
        va='bottom'
    )

plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Monthly Sales Trends Over the Past Year')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

<div style="text-align: left; background-color:black; padding: 5px; color: white; font-size: 18px; margin-top: 10px;">
    
Identify any seasonal patterns in the sales data.
</div>

 

### We see that sales go up a lot in July every year, which might mean people need our products more then. Sales tend to drop in August, which could mean it's a slower month for us.
